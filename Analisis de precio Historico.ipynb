Analisis de la fluctuacion del precio de las propiedas en los barrios de Belgrano y Villa Lugano

Tomamos los datos recuperados en los notebooks anteriores sobre las casas



```python

#Armamos nuestro dataframe
#Barrio Belgrano
#2014 1.011417e+06
#2015 1.078500e+06
#2016 1.279074e+06
#2017 9.853293e+05

import datetime as datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%config IPCompleter.greedy=True
%matplotlib inline
price_by_year= {2014: 1.011417e+06, 2015: 1.078500e+06, 2016: 1.279074e+06, 2017: 9.853293e+05}
price_year= pd.Series(price_by_year);
price_year.plot(linewidth=2,figsize=(12,8))



```

```python

#Armamos nuestro dataframe
#Barrio Villa Lugano
#2014 1.560000e+05
#2015 2.000000e+05
#2016 2.683333e+05
#2017 2.104474e+05

import datetime as datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%config IPCompleter.greedy=True
%matplotlib inline
price_by_year= {2014: 1.560000e+05, 2015: 2.000000e+05, 2016: 2.683333e+05, 2017: 2.104474e+05}
price_year= pd.Series(price_by_year);
price_year.plot(linewidth=2,figsize=(12,8))

```

Asi podemos ver que para tanto uno de los barrios mas caros como para uno de los barrios mas baratos, el precio de las casas sigue una distribucion similar, alcanznado un pico en el 2016 y descendiendo este mismo año
