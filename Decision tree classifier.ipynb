```python
import numpy as np
import pandas as pd
from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
properties = pd.read_csv('trainClusterPorPrecio800.csv')
testing = pd.read_csv('testTree.csv')
```

```python
train = properties
train = train.drop('id', 1)
train = train.drop('description', 1)
train = train.drop('Unnamed: 0', 1)
train = train.drop('Unnamed: 0.1', 1)
train = train.drop('created_on', 1)
train = train.drop('price_aprox_usd', 1)
train['lat'] = train['lat']*2000
train['lon'] = train['lon']*2000
train.head()
```

```python
test = testing
test = test.drop('description',1)
test = test.drop('Unnamed: 0',1)
test = test.drop('created_on',1)
test['lat'] = test['lat']*2000
test['lon'] = test['lon']*2000
test.head()
```

```python
trainX = train.values[:, 0:62]
trainY = train.values[:,62]
```

```python
testX = test.values[:, 1:63]
```

```python
clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100, max_depth=None, min_samples_leaf=5)
clf_gini.fit(trainX, trainY)
```

```python
#menos preciso que gini
#clf_entropy = DecisionTreeClassifier(criterion = "entropy", random_state = 100, max_depth=None, min_samples_leaf=5)
#clf_entropy.fit(trainX, trainY)
```

```python
predicciones = clf_gini.predict(testX)
predicciones
```

```python
testing['cluster']= predicciones
```
