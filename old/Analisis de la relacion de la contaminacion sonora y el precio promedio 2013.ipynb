```python
import pandas as pd
import numpy as np
#Comienzo haciendo un filtro de las propiedades segun capital federal y Gran Buenos Aires
agosto_2013 = pd.read_csv('properati-AR-2013-08-01-properties-sell.csv', low_memory = False)
septiembre_2013 = pd.read_csv('properati-AR-2013-09-01-properties-sell.csv', low_memory = False)
octubre_2013 = pd.read_csv('properati-AR-2013-10-01-properties-sell.csv', low_memory = False)
noviembre_2013 = pd.read_csv('properati-AR-2013-11-01-properties-sell.csv', low_memory = False)
diciembre_2013 = pd.read_csv('properati-AR-2013-12-01-properties-sell.csv', low_memory = False)
agosto_2013
```

```python
#Hago un merge para tener todas las propiedades de 2013
propiedades_2013 = pd.concat([agosto_2013,septiembre_2013,octubre_2013,noviembre_2013,diciembre_2013])
propiedades_2013 = propiedades_2013[propiedades_2013.created_on.str.contains("2013")].copy()
#Se tienen 306102 registros en total
```

```python
#Como solo tengo informacion de la contaminacion sonora en capital federal, me quedo con las propiedades de alli
propiedades_capital_2013 = propiedades_2013[propiedades_2013.place_with_parent_names.str.contains("Capital Federal")].copy()
propiedades_capital_2013
#Hay 96816 registros en capital federal
```

```python
propiedades_capital_2013.info()
```

```python
#Me interesa solo el precio por metro cuadrado en dolares, barrio(place_name)
propiedades_barrios_2013 = propiedades_capital_2013[['place_name','price_usd_per_m2']]
#Se tienen los precios por metro cuadrado por barrio, pero capital federal no es un barrio, asi que lo remuevo
propiedades_barrios_2013 = propiedades_barrios_2013[propiedades_barrios_2013['place_name'] != 'Capital Federal']
propiedades_barrios_2013
```

```python
#No me interesan los datos con nulos
propiedades_barrios_2013 = propiedades_barrios_2013.dropna()
propiedades_barrios_2013
#Sobreviven 72972 registros
```

```python
#Antes de agrupar por barrio, primero le cambio el nombre al atributo, para el join posterior con los datos de contaminacion
propiedades_barrios_2013['BARRIO'] = propiedades_barrios_2013['place_name']
propiedades_barrios_2013 = propiedades_barrios_2013[['BARRIO','price_usd_per_m2']]
#Agrupo por barrios
promedio_por_barrios = propiedades_barrios_2013.groupby('BARRIO').mean()
promedio_por_barrios = promedio_por_barrios.sort_values(by='price_usd_per_m2',ascending=False)
promedio_por_barrios
```

```python
promedio_por_barrios.dropna()
```

```python
#Ahora comienzo a analizar lo que tengo de contaminacion sonora en 2013
contaminacion_sonora = pd.read_csv('contaminacion acustica 2013.csv', sep = ";",low_memory = False) 
contaminacion_sonora.info()
#TMI es el numero de la torre de monitoreo, PROMEDIO_ENERGETICO_HORA representa los decibeles medidos
```

```python
#Cambio el nombre de la columna para que sea mas intuitivo
contaminacion_sonora['dB'] = contaminacion_sonora['PROMEDIO_ENERGETICO_HORA']
contaminacion_sonora = contaminacion_sonora[['TMI','FECHA','dB']]
contaminacion_sonora[:5]#En total hay 91584 registros
```

```python
#Ahora veo que informacion me aportan las torres de monitoreo
df_torres = pd.read_csv('torres-de-monitoreo-inteligente.csv', low_memory = False)
df_torres.info()
```

```python
df_torres#Hay datos de 16 torres en total(Una por barrio, osea que tengo datos de 16 barrios en capital federal)
```

```python
#Hago un merge de la informacion de las torres y los decibeles medidos por barrio
contaminacion_sonora_barrios = pd.merge(contaminacion_sonora, df_torres, on='TMI', how = 'inner')
contaminacion_sonora_barrios = contaminacion_sonora_barrios[['TMI','FECHA','dB','BARRIO']]
#81408 registros en total
```

```python
#Agrupo por barrios
contaminacion_promedio =  contaminacion_sonora_barrios.groupby('BARRIO').mean()
contaminacion_promedio =  contaminacion_promedio.sort_values(by='dB', ascending=False)
dB_promedio = contaminacion_promedio['dB']
dB_promedio
```

```python
%matplotlib inline
dB_promedio.plot(kind = 'bar', title = 'Contaminacion promedio barrio')
```

```python
dB_promedio_index = dB_promedio.reset_index()
promedio_por_barrios_index = promedio_por_barrios.reset_index()
#Ahora hago un merge de los datos de contaminacion y precio por barri
precios_y_contaminacion = pd.merge(dB_promedio_index,propiedades_barrios_2013,on='BARRIO', how = 'inner')
precios_y_contaminacion =  precios_y_contaminacion[['BARRIO','dB','price_usd_per_m2']]
precios_y_contaminacion
```

```python
#Hago un scatter plot para ver si hay alguna tendencia que sugiera que hay relacion entre las magnitudes
precios_y_contaminacion = precios_y_contaminacion[precios_y_contaminacion['price_usd_per_m2']<5000]
precios_y_contaminacion.plot.scatter('dB','price_usd_per_m2',alpha=0.25,figsize=(12,8),title="Relacion contaminacion sonora y precio");
```
